```python
import numpy
import math
from matplotlib import pyplot
# embed the figures into the notebook
%matplotlib inline
%run ./definitions.ipynb
```

```python
n = 50
xa, xz = -2, 2
ya, yz = -1, 1

x,y,X,Y = makemesh(n,xa,xz,ya,yz)
```

```python
gamma = 5
xvort, yvort = 0,0
```

```python
uvort, vvort = getvelvortex(gamma,xvort,yvort,X,Y)
psivort = getstreamfuncvortex(gamma,xvort,yvort,X,Y)
```

```python
# plot the streamlines
size = 10
pyplot.figure(figsize=(size, (yz-ya)/(xz-xa)*size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(xa, xz)
pyplot.ylim(ya, yz)
pyplot.streamplot(X, Y, uvort, vvort, density=2, linewidth=1, arrowsize=1, arrowstyle='->',color='#415A77')
pyplot.scatter(xvort, yvort, color='#FB3640', s=80, marker='o');
```

```python
sinkstr = -1
xsink, ysink = 0,0

usink, vsink = getvel(sinkstr,xsink,ysink,X,Y)
psisink = getstreamfunc(sinkstr,xsink,ysink,X,Y)
```

```python
# superposition of the sink and the vortex
u = uvort + usink
v = vvort + vsink
psi = psivort + psisink

# plot the streamlines
size = 10
pyplot.figure(figsize=(size, (yz-ya)/(xz-xa)*size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(xa, xz)
pyplot.ylim(ya, yz)
pyplot.streamplot(X, Y, u, v, density=2, linewidth=1, arrowsize=1, arrowstyle='->',color='#415A77')
pyplot.scatter(xvort, yvort, color='blue', s=80, marker='o');
```
