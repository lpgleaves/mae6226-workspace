```python
N = 547
xv = numpy.linspace(-109,109,N)
yv = numpy.zeros(N)

# creating my mesh grid X,Y
n = 50
xa, xz = -2, 2
ya, yz = -0.5,0.5

x,y,X,Y = makemesh(n,xa,xz,ya,yz)
```

```python
ucomp = 0
vcomp = 0
gamma = 10

for i in range(0,N-1):
    u,v = getvelvortex(gamma,xv[i],yv[i],X,Y)
    ucomp = ucomp + u
    vcomp = vcomp + v
 
```

```python
# plot the streamlines
size = 15
pyplot.figure(figsize=(size, (yz-ya)/(xz-xa)*size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(xa, xz)
pyplot.ylim(ya,yz)
pyplot.streamplot(X, Y, ucomp, vcomp, density=1, linewidth=1, arrowsize=1, arrowstyle='->')
pyplot.scatter(xv, yv, s=80, marker='o', color='#B22222');
```
